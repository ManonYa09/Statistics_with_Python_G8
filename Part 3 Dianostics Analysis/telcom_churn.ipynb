```python
customer = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']
fig , axes = plt.subplots(2, 2, figsize= (12, 6))
for i, ax in enumerate(axes.flat):
    sns.countplot(data = df,x =  customer[i], hue = df['Churn'], ax = ax, width=0.6)
    ax.set_title(f'{customer[i]}')
    ax.set_ylabel('Count')
    for i in ax.containers:
        ax.bar_label(i)
sns.despine()
plt.tight_layout()    
```
